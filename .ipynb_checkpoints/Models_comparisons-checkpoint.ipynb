### 1. Load Datasets

```python
with_eur = False
early = True
train_test_ratio = 0.7
if early:
    START_DATE = dt.datetime(1960, 2, 1)
elif with_eur:
    START_DATE = dt.datetime(2001, 2, 1)
else:
    START_DATE = dt.datetime(1992, 1, 1)
END_DATE = dt.datetime.today() - pd.offsets.MonthEnd(2)
TEST_DATE = START_DATE + train_test_ratio * (END_DATE - START_DATE)
ROW_SHIFTS = (1, 2, 3, 6, 12)
predicted_feature = 'Fed_rate_month_avg_diff'
# predicted_feature = 'Fed_rate_month_avg_trend'
# predicted_feature = 'Fed_rate_month_avg_diff_5_class'

path = 'Models/startdate_{:s}_testdate_{:s}/'.format(START_DATE.strftime("%Y"),
                                                     TEST_DATE.strftime("%Y")
                                                    )
file_names = [f + "_t-" + "_".join([str(i) for i in ROW_SHIFTS]) + ".csv" for f in ["X_train", "Y_train", "X_test", "Y_test"]]

X_train, Y_train, X_test, Y_test = open_files(path=path,
                                              file_names=file_names
                                             ).values()
Y_train['Date'] = pd.to_datetime(Y_train['Date'])
Y_test['Date'] = pd.to_datetime(Y_test['Date'])
Y_test.dtypes
```

### 2. Load models

```python
models = open_files(path=path,
                    file_names=get_file_names(path=path,
                                              extension='.pkl'
                                             )
                   )
print(' available files '.center(120, "-"))
print("\n".join(models.keys()))
models = {key: value for (key, value) in models.items() if predicted_feature.replace("_", "") in key}
print(" Loaded models ".center(120, "-"))
print("\n".join(models.keys()))
models = list(models.values())
# models.append(LinearDiscriminantAnalysis().fit(X_train, Y_train[predicted_feature]))
# models.append(LogisticRegression().fit(X_train, Y_train[predicted_feature]))
# models.append(RandomForestClassifier().fit(X_train, Y_train[predicted_feature]))
# models.append(GradientBoostingClassifier().fit(X_train, Y_train[predicted_feature]))
# models.append(XGBClassifier().fit(X_train, Y_train[predicted_feature]))

models.append(LinearRegression().fit(X_train, Y_train[predicted_feature]))
models.append(RandomForestRegressor().fit(X_train, Y_train[predicted_feature]))
models.append(GradientBoostingRegressor().fit(X_train, Y_train[predicted_feature]))
models.append(XGBRegressor().fit(X_train, Y_train[predicted_feature]))
```

```python
model_params = []
for m in models:
    m_name = m.__class__.__name__
    if m_name == "GradientBoostingRegressor" :
        default_params = GradientBoostingRegressor().get_params().items()
        params = dict(set(m.get_params().items()) - set(default_params))
    elif m_name == "RandomForestRegressor":
        default_params = RandomForestRegressor().get_params().items()
        params = dict(set(m.get_params().items()) - set(default_params))
    elif m_name == "RandomForestClassifier":
        default_params = RandomForestClassifier().get_params().items()
        params = dict(set(m.get_params().items()) - set(default_params))
        params = {key: value for (key, value) in params.items() if key in ['n_estimators', 'max_features', 'min_samples_leaf']}
    elif m_name =="GradientBoostingClassifier":
        default_params = GradientBoostingRegressor().get_params().items()
        params = dict(set(m.get_params().items()) - set(default_params))
        params = {key: value for (key, value) in params.items() if key in ['n_estimators', 'max_depth', 'learning_rate', 'ccp_alpha', 'min_samples_leaf', 'max_features']}
    elif m_name == "LinearRegression":
        default_params = LinearRegression().get_params().items()
        params = dict(set(m.get_params().items()) - set(default_params))
    elif m_name == "XGBRegressor":
        default_params = XGBRegressor().get_params().items()
        params = dict(set(m.get_params().items()) - set(default_params))
        params = {key: value for (key, value) in params.items() if key in ['n_estimators', 'max_depth', 'learning_rate']}
    try:
        del params['random_state']
    except: pass        
    model_params.append(params)
```

### 3. Plot models

```python
def feature_importance(fitted_model):
    try:
        # lack an if statement with isinstance linear reg, boosting, rf, etc
        importance = fitted_model.feature_importances_
        std = np.std([tree.feature_importances_ for tree in fitted_model.estimators_], axis=0)
        idx = np.argsort(importance)[::-1]
        plt.style.use("seaborn-darkgrid")
        plt.figure(figsize = (10, 0.2 * len(idx)))
        sns.barplot(x=importance[idx], y=X_train.columns[idx], color=sns.color_palette()[0])
        plt.title("Feature importance for current model")
        plt.show()
    except: pass
```
